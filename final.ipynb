```python
def aqifind(List):
    Aqi = List[0]
    AQI_Bucket = List[1]
    
    if pd.isnull(AQI_Bucket):
        if 0<=Aqi<=50:
            return 0
        elif 51<=Aqi<=100:
            return 1
        elif 101<=Aqi<=200:
            return 2
        elif 201<=Aqi<=300:
            return 3
        elif 301<=Aqi<=400:
            return 4
        elif 401<=Aqi:
            return 5
    else:
        if AQI_Bucket=="Good":
            return 0
        elif AQI_Bucket=="Satisfactory":
            return 1
        elif AQI_Bucket=="Moderate":
            return 2
        elif AQI_Bucket=="Poor":
            return 3
        elif AQI_Bucket=="Very Poor":
            return 4
        elif AQI_Bucket=="Severe":
            return 5
```

For loop for finding optimal K value for given data . (Found to be 15) Generally they assume best k is sqrt of N so i ran a for loop till sqrt of N that is 171 and found maximum accurracy at k=15

```python
accuracy = []
for i in range(1,171):
    knn = neighbors.KNeighborsClassifier(n_neighbors=i,weights="distance")
    knn.fit(x_train1,y_train1)
    predictions = knn.predict(x_test1)
    Accu = accuracy_score(y_test1,predictions)
    accuracy.append(Accu)
```

```python
plt.figure(figsize=(10,6))

plt.plot(range(1,171),accuracy,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
        
plt.title('Accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Accuracy",max(accuracy),"at K =",accuracy.index(max(accuracy)))
```

POLYNOMIAL REGRESSION

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
train_rmse_errors = []
test_rmse_errors = []
```

CHECKING OPTIMAL VALUE OF DEGREE OF POLYNOMIAL SUCH THAT THE MODEL DOES NOT UNDERFIT OR OVERFIT

```python


for d in range(1,4):
  poly_conv = PolynomialFeatures(degree=d,include_bias=False)
  poly_fea = poly_conv.fit_transform(dfpolyx)
  x_train2,x_test2,y_train2,y_test2=train_test_split(poly_fea,dfpolyy,test_size=0.3,random_state=101)
  model = LinearRegression()
  model.fit(x_train2,y_train2)
  train_pred = model.predict(x_train2)
  test_pred = model.predict(x_test2)
  rmse1 = np.sqrt(mean_squared_error(y_train2,train_pred))
  rmse2 = np.sqrt(mean_squared_error(y_test2,test_pred))
  train_rmse_errors.append(rmse1)
  test_rmse_errors.append(rmse2)

```

```python
test_rmse_errors
```

```python
plt.plot(range(1,4),train_rmse_errors,label="Train")
plt.plot(range(1,4),test_rmse_errors,label="Test")
plt.ylabel("RMSE")
plt.xlabel("Degree")
plt.legend()
```
